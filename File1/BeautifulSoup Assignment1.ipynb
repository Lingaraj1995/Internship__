```python
#Question7
restaurant=requests.get('https://www.dineout.co.in/mumbai-restaurants/welcome-back')
restaurant
```

```python
restaurant=BeautifulSoup(restaurant.content)
restaurant
```

```python
restaurant_names=[]

for i in restaurant.find_all('div',class_="restnt-info cursor"):
    restaurant_names.append(i.text)
    
restaurant_names
```

```python
cuisine=[]

for i in restaurant.find_all('span',class_="double-line-ellipsis"):
    cuisine.append(i.text.replace('₹',''))
    
cuisine
```

```python
location=[]

for i in restaurant.find_all('div',class_="restnt-loc ellipsis"):
    location.append(i.text)
    
location
```

```python
ratings=[]

for i in restaurant.find_all('div',class_="restnt-rating"):
    ratings.append(i)
    
ratings
```

```python
images=[]

for i in restaurant.find_all('img',class_="no-img"):
    images.append(i['data-src'])
    
images
```

```python
restaurantdf=pd.DataFrame({'Restaurant Names':restaurant_names,'cusine':cuisine,'location':location,'ratings':ratings,'image_url':images})
```
