```python
import time
```

```python
data=[]

for i in driver.find_elements(By.XPATH,'//a[@class="link-hover link text-xl font-semibold"]'):
    next_button=driver.find_elements(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[3]/div/button[2]')
    time.sleep(2)

data
```

```python
data_urls=[]

for page in range(0,10):
    url=driver.find_elements(By.XPATH,'//a[@class="link-hover link text-xl font-semibold"]')
    for i in url:
        data_urls.append(i.get_attribute('href')) 
    next_button=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[3]/div/button[2]')
    next_button.click()
    time.sleep(2)
    
    
data_urls
```

```python
data_urls
```

```python
len(data_urls)
```

```python
Dataset_name=[]

for i in data_urls:
    driver.get(i)
    name=driver.find_element(By.XPATH,'//div[@class="relative flex flex w-full items-center gap-4 bg-primary p-2"]/div/div/h1')
    Dataset_name.append(name.text)
    
Dataset_name
```

```python
len(Dataset_name)
```

```python
Data_type=[]

for i in data_urls:
    driver.get(i)
    Type=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[4]/p')
    Data_type.append(Type.text)
    
Data_type
```

```python
Task=[]

for i in data_urls:
    driver.get(i)
    task=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[3]/p')
    Task.append(task.text)
    
Task
```

```python
Instances=[]

for i in data_urls:
    driver.get(i)
    instance=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[5]/p')
    Instances.append(instance.text)

Instances
```

```python
Year=[]

for i in data_urls:
    driver.get(i)
    try:
        year=driver.find_element(By.XPATH,'//h2[@class="text-sm text-primary-content"]')
        Year.append(year.text)
    except NoSuchElementException:
        Year.append("-")
    
    
Year
```
